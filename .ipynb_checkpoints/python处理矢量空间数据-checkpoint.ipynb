```python
from osgeo import ogr
import requests
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']  # 设置字体为微软雅黑
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号
```

```python
url = 'https://geo.datav.aliyun.com/areas_v3/bound/410000_full.json'

try:
    r = requests.get(url, timeout=30)
    status = r.raise_for_status()
    r.encoding = r.apparent_encoding
    print(r.status_code)
#     print(requests.head(url).headers)
#     print(r.text[0:1000])
 
except:
    print("error!")
```

```python
ds = ogr.Open(r.text)
# print(type(ds))
```

```python
lyr = ds.GetLayer()
# print(lyr.GetFeatureCount())
for fea in lyr:
    name = fea.GetField('name')
#     print("name",name)
    geom = fea.geometry()
    ring1 = geom.GetGeometryRef(0)
    ring2 = ring1.GetGeometryRef(0)

    coords = ring2.GetPoints()
    x,y = zip(*coords)
    plt.plot(x,y,'black')
#     plt.fill(x,y,'b')
    i=x[0]
    j=y[0]
    plt.text(i,j,name,color='red',fontsize=12,ha='center',va='center')
plt.axis('equal')
plt.show()
    
```
